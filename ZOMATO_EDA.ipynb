# 1. Finding Missing Values

```python
df.isnull().sum()
```

```python
# getting columns which have null values
null_cols = []
for col in df.columns :
    if df[col].isnull().sum() > 0:
        null_cols.append(col)
null_cols
```

```python
df_country = pd.read_excel("Country-Code.xlsx")
df_country.head()
```

```python
# 'Country Code' is also present in zomato dataframe so merging both the dataframes
final_df  = pd.merge(df,df_country,on="Country Code",how='left')
```

```python
final_df.head()
```

```python
# checking data types
final_df.dtypes
```

```python
# finding records of each country in the dataset
final_df.Country.value_counts()
```

```python
country_names  = final_df.Country.value_counts().index
country_values = final_df.Country.value_counts().values
```

```python
# pie chart representation for top 3 countries that use zomato
plt.pie(country_values[:3],labels=country_names[:3],autopct="%1.2f%%")
```

# Analysis :
    Zomato's maximm records or trasactions are from india,
    after that USA and United Kingdom 

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

# Analysis :
    1. when ratings are between 4.5 to 4.0 --> Excellent
    2. when ratings are between 4.0 to 4.4 --> very good
    3. when ratings are between 3.5 to 3.9 --> good
    4. when ratings are between 3.0 to 3.4 --> average
    5. when ratings are between 2.5 to 2.9 --> average
    6. when ratings are between 1.8 to 2.4 --> poor
    7. when customers do not rate the food its given 0 ratings

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating", y= "Rating Count",
            hue='Rating color',data=ratings,
            palette=['white','red','orange','yellow','green','green']
           )
```

# Analysis:
    1. not rated count is very high
    2. maximum number of ratings are between 2.5 to 3.4

```python
## Count plot
sns.countplot(x="Rating color",data=ratings,palette=['white','red','orange','yellow','green','green'])
```

```python
# countries that have given 0 rating
final_df[final_df['Rating color'] == 'White'].groupby('Country').size().reset_index()
```
